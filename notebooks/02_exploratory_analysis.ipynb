# Exploratory Data Analysis (EDA)
Visualizing trends, distributions, and identifying patterns in order data.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/cleaned/orders_cleaned.csv', parse_dates=['order_date','ship_date','delivery_date'])
```

```python
# Distribution of order to delivery days
plt.hist(df['order_to_delivery_days'].dropna(), bins=20, color='skyblue')
plt.xlabel('Order to Delivery Days')
plt.ylabel('Number of Orders')
plt.title('Distribution of Delivery Times')
plt.show()
```

```python
# Count of late deliveries by region
late_by_region = df[df['is_late']].groupby('region').size()
late_by_region.plot(kind='bar', color='salmon')
plt.ylabel('Number of Late Orders')
plt.title('Late Deliveries by Region')
plt.show()
```

```python
# Top 5 delay reasons
df['delay_reason'].value_counts().head(5).plot(kind='bar', color='orange')
plt.ylabel('Count')
plt.title('Top 5 Delay Reasons')
plt.show()
```
